# m=MATPLOTLIB 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Basic Graph

```python
x=[1,2,3]
y=[2,4,6]
x2= np.arange(0,4.5,0.5)
plt.plot(x,y, label='2x',color='green', linewidth=1.5,marker='*')

plt.title('Basic Graph', fontdict={'fontname':'Times New Roman','fontsize':20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.plot(x2,x2**2,'r*--')

plt.xticks([1,2,3,4])
#plt.yticks([0,1,2,3,4,5,6])

plt.legend()

plt.show()

```

## Histogram

```python
#Reading the titanic dataset

df=pd.read_csv('https://storage.googleapis.com/tf-datasets/titanic/train.csv')
df.head()
#Checking the ratio of male to female
x=df['sex']
plt.title('SEX ')
x.hist(label='2x',color='green')
plt.xlabel('gender')
plt.ylabel('number')

plt.show()
```

## Bar Plot

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
langs=['C', 'C++', 'Java', 'Python', 'PHP']
students=[23,17,35,29,12]
ax.bar(langs,students)
plt.show()
```

## Pie Chart

```python

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.axis('equal')
langs=['C', 'C++', 'Java', 'Python', 'PHP']
students=[23,17,35,29,12]
ax.pie(students, labels=langs,autopct='%1.2f%%')
plt.show()
```
